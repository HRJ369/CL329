```python
import matplotlib.pyplot as plt
import control as ctrl
import numpy as np

# Define the range of Kc values for the root locus plot
Kc_values = np.linspace(0, 40, 100)  # Adjust the range as needed

# Define the open-loop transfer function G(s) = 1 / (6s^2 + 5s + 1)
num = [1]
den = [6, 5, 1]

# Create the transfer function without Kc
G = ctrl.TransferFunction(num, den)

# Plot the root locus for the system with varying Kc
plt.figure()
ctrl.root_locus(G, kvect=Kc_values, grid=True)
plt.title("Root Locus Diagram")
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.show()

```
